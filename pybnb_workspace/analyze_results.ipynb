```python
import pandas as pd
import numpy as np

def print_full(df):
    # Set display options to show all columns
    pd.set_option('display.max_columns', None)  # Show all columns
    pd.set_option('display.max_rows', None)     # Show all rows
    pd.set_option('display.width', None)        # Auto-detect display width
    pd.set_option('display.max_colwidth', None) # Show full content of each column
    display(df)
    # Reset display options to default
    pd.reset_option('display.max_columns')
    pd.reset_option('display.max_rows')
    pd.reset_option('display.width')
    pd.reset_option('display.max_colwidth')
```

```python
results_id = {
    "mat_type_a_1":"20250904-152518",
    "mat_type_b_1":"20250904-152058",
    "mat_type_a_2":"20250906-173524",
    "mat_type_b_2":"20250906-172413",
    "mat_type_a_3":"20250908-175440",
    "mat_type_b_3":"20250908-175725",
}
results_df = {}
```

```python
for key, val in results_id.items():
    results_df[key] = pd.read_csv(f"./../Modules/output_processor/processed_results_{val}.csv")
    results_df[key][['instance_type', 'instance_id']] = results_df[key][['instance_type', 'instance_id']].astype(float)
    print(f"{key}: {results_df[key].shape}")
```

```python
cmp_inst = [1,2,3]
ref_columns_base = ['instance_type','instance_id']
ref_columns = [f"{r}_{i}" for r in ref_columns_base for i in cmp_inst]
comparing_columns_base = ['bnb_upb', 'bnb_lwb',
                     'root_node_relaxObj', 'bnb_termCond',
                     "bnb_nodesExp", "bnb_gap", "bnb_wallTime"]
comparing_columns = [f"{c}_{i}" for c in comparing_columns_base for i in cmp_inst]

agg_df = pd.merge(pd.merge(results_df['mat_type_a_1'].add_suffix('_1', axis=1), results_df['mat_type_a_2'].add_suffix('_2', axis=1),how='inner',
         left_index=True, right_index=True), results_df['mat_type_a_3'].add_suffix('_3', axis=1), how='inner',
         left_index=True, right_index=True)[ref_columns+comparing_columns]
print_full(agg_df)
print_full(agg_df.loc[lambda x: x['bnb_termCond_3']!=x['bnb_termCond_2']])
```

```python
cmp_inst = [1,2,3]
ref_columns_base = ['instance_type','instance_id']
ref_columns = [f"{r}_{i}" for r in ref_columns_base for i in cmp_inst]
comparing_columns_base = ['bnb_upb', 'bnb_lwb',
                     'root_node_relaxObj', 'bnb_termCond',
                     "bnb_nodesExp", "bnb_gap", "bnb_wallTime"]
comparing_columns = [f"{c}_{i}" for c in comparing_columns_base for i in cmp_inst]

print_full(pd.merge(pd.merge(results_df['mat_type_b_1'].add_suffix('_1', axis=1), results_df['mat_type_b_2'].add_suffix('_2', axis=1),how='inner',
         left_index=True, right_index=True), results_df['mat_type_b_3'].add_suffix('_3', axis=1), how='inner',
         left_index=True, right_index=True)[ref_columns+comparing_columns])
```
